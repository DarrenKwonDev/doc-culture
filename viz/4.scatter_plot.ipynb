## scatter

- 전반적인 분포, 상관관계
- scatter plot의 목적 : 군집, 값 사이의 차이, 이상치  
- 점이 많아지면 힘듦
  - 투명도 조정, jittering, 히트맵, contour plot(등고선) 정도로 해결하려고 함
- 추세선 그리는 건 좋은데 한 개만 쓰세요. 가독성을 떨어뜨림.
- grid 그리지마. 사용한다면 넓직하게 그려라.  
- 솔직히 seaborn에서 그리는 게 훨씬 편해!

```python
fig, ax = plt.subplots(figsize=(10, 6))
x = np.random.rand(20)
y = np.random.rand(20)

ax.scatter(x, y, 
    s=100, # size
    c="royalblue", # color
    marker="o", 
    alpha=0.5, 
    edgecolors="black", 
    linewidths=0.5)
```
