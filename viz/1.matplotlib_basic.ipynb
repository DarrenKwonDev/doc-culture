## 기본 plot

figure와 axes 객체

```python
fig = plt.figure() # fig = figure
plt.show() # <Figure size 640x480 with 0 Axes>
```

```python
fig = plt.figure()
ax = fig.add_subplot() # Add an Axes to the figure as part of a subplot arrangement.
plt.show()
```

```python
fig = plt.figure(figsize=(12, 7)) # 단위는 inch
ax = fig.add_subplot()
plt.show()
```

```python
fig = plt.figure()

# https://matplotlib.org/stable/api/figure_api.html#matplotlib.figure.Figure.add_subplot
# (nrows, ncols, index). default: (1, 1, 1)

ax = fig.add_subplot(1, 2, 1) # 즉, col을 2개로 나누고 첫번째에 그린다.
ax = fig.add_subplot(1, 2, 2) # 즉, col을 2개로 나누고 두번째에 그린다.

plt.show()
```

```python
fig = plt.figure()

# 2 x 2로 쪼개고 2, 3번째 인덱스에 배치한다.
ax = fig.add_subplot(2, 2, 2)
ax = fig.add_subplot(2, 2, 3)

plt.show()
```

axes.plot, axes.bar

```python
fig = plt.figure()
ax = fig.add_subplot()

x = np.array([1, 5, 100, 2])

plt.plot(x) # https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot
plt.show()
```

```python
fig = plt.figure()

x1 = np.array([1, 2, 3])
x2 = np.array([3, 2, 1])

# plot으로 데이터를 그리면 가장 가깝게 선언된 axes에 그려짐

ax1 = fig.add_subplot(211)
plt.plot(x1) # ax1에 그리기

ax2 = fig.add_subplot(2, 1, 2)
plt.plot(x2) # ax2에 그리기

plt.show()
```

```python
fig = plt.figure()

x1 = np.array([1, 2, 3])
x2 = np.array([3, 2, 1])

# 이 방법이 훨씬 편함
ax1 = fig.add_subplot(2, 1, 1)
ax1.plot(x1) # ax1에 그리기
ax2 = fig.add_subplot(2, 1, 2)
ax2.plot(x2)

plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot()

# 자동 색 구분
ax.plot([1, 1, 1])
ax.plot([1, 2, 3])
ax.plot([3, 3, 3])
```

```python
fig = plt.figure()
ax = fig.add_subplot()

ax.plot([1, 2, 3], [1, 2, 3], color="red")
ax.bar([1, 2, 3], [1, 2, 3], color="blue")
```

plt.legend, plt.title, plt.xlabel, plt.ylabel

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot([1, 1, 1], label="1")
ax.plot([1, 2, 3], label="2")
ax.plot([3, 3, 3], label="3")
plt.legend() # 레전드도 보여줘
plt.title("title") # 타이틀도 보여줘
plt.xlabel("x label") # x축 레이블도 보여줘
plt.ylabel("y label") # y축 레이블도 보여줘
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)

ax1.set_title("ax1")
ax2.set_title("ax2")
fig.suptitle("fig") # fig의 타이틀. (super title의 준말)
```

axes.ticks, axes.ticklabels

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot([1, 1, 1], label="1")
ax.plot([1, 2, 3], label="2")
ax.plot([3, 3, 3], label="3")

ax.legend()

ax.set_xticks([0, 1, 2])
ax.set_xticklabels(["zero", "one", "two"])
```

axes.text, axes.annotate

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot([1, 1, 1], label="1")
ax.plot([1, 2, 3], label="2")
ax.plot([3, 3, 3], label="3")

ax.legend()

ax.text(x=1, y=2, s="text") # (x, y)에 텍스트를 그린다.
ax.annotate("annotate", xy=(1, 1)) # (x, y)에 화살표를 그린다.
```
